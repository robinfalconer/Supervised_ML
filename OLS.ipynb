```python
import statsmodels.api as sm
import pandas as pd
```

```python
df = pd.read_csv("wind_farm")
df.head()
```

```python
df.columns
```

```python
target_var = 'Energy Export (kWh)'
X = df.drop(columns=[target_var, 'Power (kW)', 'Quarter'])
y = df[target_var]
```

```python
X = pd.get_dummies(X, columns=['Month','wind_cat','wind_direction_cat'], drop_first=True)
```

```python
X = sm.add_constant(X)
```

```python
X.columns
```

```python
#X.drop(columns=['Wind direction (°)','Nacelle position (°)','Production Factor','Blade angle (pitch) (°)'], inplace=True)
#X.drop(columns=['Nacelle position in relation to wind direction (°)'], inplace=True)
X.drop(columns=['Wind speed (m/s)'], inplace=True)
```

```python
model = sm.OLS(y, X)
result = model.fit()
result.summary()
```
