```python
import pandas as pd
df = pd.read_csv("wind_farm")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True)
```

```python
df.groupby(df.Month).mean()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(df["Blade angle (pitch) (°)"])
plt.show()
```

```python
quarter_wind = df.groupby(df["Quarter"]).mean()["Wind speed (m/s)"]

plt.figure(figsize=(12,4))
sns.kdeplot(df["Wind speed (m/s)"], hue=df["Quarter"], palette="Set2")
plt.title(f"Average wind speed:\nQ1: {round(quarter_wind[1],1)} m/s  |   Q2: {round(quarter_wind[2],1)} m/s  |   Q3: {round(quarter_wind[3],1)} m/s  |   Q4: {round(quarter_wind[4],1)} m/s",
         fontsize=15)
plt.xlabel("Wind speed (m/s)",fontsize=15)
plt.ylabel("")
plt.legend(["Q1","Q2","Q3","Q4"], fontsize=14)
plt.show()
```

```python
quarter_temp = df.groupby(df["Quarter"]).mean()["Ambient temperature (°C)"]

plt.figure(figsize=(12,4))
sns.kdeplot(df["Ambient temperature (°C)"], hue=df["Quarter"], palette="Set2")
plt.title(f"Average temperature:\nQ1: {round(quarter_temp[1],1)}°C  |   Q2: {round(quarter_temp[2],1)}°C  |   Q3: {round(quarter_temp[3],1)}°C  |   Q4: {round(quarter_temp[4],1)}°C",
         fontsize=15)
plt.xlabel("Ambient temperature (°C)",fontsize=15)
plt.ylabel("")
plt.legend(["Q1","Q2","Q3","Q4"], fontsize=14)
plt.show()
```

```python
wind_cat_kwh = df.groupby(df["wind_cat"]).mean()["Energy Export (kWh)"]

plt.figure(figsize=(12,8))
sns.kdeplot(df["Energy Export (kWh)"], hue=df["wind_cat"], palette="Set3")
#plt.title(f"Average wind speed\nQ1: {round(wind_cat_kwh[1],1)} m/s\nQ2: {round(wind_cat_kwh[2],1)} m/s\nQ3: {round(wind_cat_kwh[3],1)} m/s\nQ4: {round(wind_cat_kwh[4],1)} m/s")
plt.show()
```

```python
df["wind_direction_cat"] = pd.cut(df["Wind direction (°)"], bins=[0,90,180,270,360], labels=["NE","SE","SW","NW"])
```

```python
df.groupby(df["wind_direction_cat"]).mean()["Wind speed (m/s)"][0]
```

```python
quarter_wind = df.groupby(df["wind_direction_cat"]).mean()["Wind speed (m/s)"]

plt.figure(figsize=(12,4))
sns.kdeplot(df["Wind speed (m/s)"], hue=df["wind_direction_cat"], palette="Set2")
plt.title(f"Average wind speed:\nNorthEast: {round(quarter_wind[0],1)} m/s  |   SouthEast: {round(quarter_wind[1],1)} m/s  |   SouthWest: {round(quarter_wind[2],1)} m/s  |   NorthWest: {round(quarter_wind[3],1)} m/s",
         fontsize=15)
plt.xlabel("Wind speed (m/s)",fontsize=15)
plt.ylabel("")
plt.legend(["North East","South East","South West","North West"], fontsize=14)
plt.show()
```

```python
quarter_temp = df.groupby(df["wind_direction_cat"]).mean()["Ambient temperature (°C)"]

plt.figure(figsize=(12,4))
sns.kdeplot(df["Ambient temperature (°C)"], hue=df["wind_direction_cat"], palette="Set2")
plt.title(f"Average temperature:\nNorthEast: {round(quarter_temp[0],1)}°C  |   SouthEast: {round(quarter_temp[1],1)}°C  |   SouthWest: {round(quarter_temp[2],1)}°C  |   NorthWest: {round(quarter_temp[3],1)}°C",
         fontsize=15)
plt.xlabel("Ambient temperature (°C)",fontsize=15)
plt.ylabel("")
plt.legend(["North East","South East","South West","North West"], fontsize=14)
plt.show()
```

```python
quarter_wind = df.groupby(df["wind_cat"]).mean()["Wind speed (m/s)"]

plt.figure(figsize=(12,8))
sns.kdeplot(df["Wind speed (m/s)"], hue=df["wind_cat"], palette="Set2")
plt.title(f"Average wind speed:\nNorthEast: {round(quarter_wind[0],1)} m/s  |   SouthEast: {round(quarter_wind[1],1)} m/s  |   SouthWest: {round(quarter_wind[2],1)} m/s  |   NorthWest: {round(quarter_wind[3],1)} m/s")
plt.show()
```
